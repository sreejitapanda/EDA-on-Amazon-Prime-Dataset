## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd

# Load datasets
credit = pd.read_csv('/content/credits.csv')
title = pd.read_csv('/content/titles.csv')

# finding shape
print("Credit Shape:", credit.shape)
print("Title Shape:", title.shape)
# find missing value

print("\nMissing Values (Credit):\n", credit.isnull().sum())
print("\nMissing Values (Title):\n", title.isnull().sum())

# drop duplicates
credit.drop_duplicates(inplace=True)
title.drop_duplicates(inplace=True)

# drop missing value
# Credit → drop rows with null values
credit.dropna(inplace=True)

# Title → drop unnecessary columns
title.drop(columns=['age_certification','seasons','tmdb_score','imdb_votes'],
           inplace=True, errors='ignore')

# Fill missing values properly
for col in title.columns:
    if title[col].dtype == 'object':
        title[col].fillna('Unknown', inplace=True)
    else:
        title[col].fillna(0, inplace=True)

print("\nData Types:\n", title.dtypes)
credit.reset_index(drop=True, inplace=True)
title.reset_index(drop=True, inplace=True)

# Merge dataset
df = pd.merge(title, credit, on='id', how='inner')

print("\nFinal Shape:", df.shape)
print("\nRemaining Missing Values:\n", df.isnull().sum())
# 8. Save Clean Dataset

df.to_csv('/content/cleaned_dataset.csv', index=False)

print("\n✅ Data Wrangling Completed Successfully!")
````

Additional operations like value_counts() and sort_values() were used to analyze popularity and ratings.
