**Data Wrangling**

```python
# Write your code to make your dataset analysis ready.
import pandas as pd

# Load datasets
credit = pd.read_csv('/content/credits.csv')
title = pd.read_csv('/content/titles.csv')

# finding shape
print("Credit Shape:", credit.shape)
print("Title Shape:", title.shape)

# find missing value

print("\nMissing Values (Credit):\n", credit.isnull().sum())
print("\nMissing Values (Title):\n", title.isnull().sum())

# drop duplicates
credit.drop_duplicates(inplace=True)
title.drop_duplicates(inplace=True)

# drop missing value
# Credit → drop rows with null values
credit.dropna(inplace=True)

# Title → drop unnecessary columns
title.drop(columns=['age_certification','seasons','tmdb_score','imdb_votes'],
           inplace=True, errors='ignore')

# Fill missing values properly
for col in title.columns:
    if title[col].dtype == 'object':
        title[col].fillna('Unknown', inplace=True)
    else:
        title[col].fillna(0, inplace=True)

print("\nData Types:\n", title.dtypes)
credit.reset_index(drop=True, inplace=True)
title.reset_index(drop=True, inplace=True)

# Merge dataset
df = pd.merge(title, credit, on='id', how='inner')

print("\nFinal Shape:", df.shape)
print("\nRemaining Missing Values:\n", df.isnull().sum())
# 8. Save Clean Dataset

df.to_csv('/content/cleaned_dataset.csv', index=False)

print("\n✅ Data Wrangling Completed Successfully!")
```

```python
# Chart - 5 visualization code
# Regional Availability: How does content distribution vary across different regions?
# Horizontal Bar chart
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/cleaned_dataset.csv')

df['production_countries'] = df['production_countries'].fillna('Unknown')

# Split multiple countries and explode
df['production_countries'] = df['production_countries'].str.split(',')
df_exploded = df.explode('production_countries')

# Count top regions
top_regions = df_exploded['production_countries'].value_counts().head(10)

# Plot horizontal bar chart
plt.figure(figsize=(10,6))
plt.barh(top_regions.index, top_regions.values)

plt.xlabel('Number of Shows')
plt.ylabel('Region')
plt.title('Top 10 Content Distribution by Region')

# Show highest on top
plt.gca().invert_yaxis()

plt.show()

```

```python

# IMDb Ratings & Popularity: What are the highest-rated or most popular shows on the platform

# Remove duplicates (same show repeated due to multiple cast)
df_unique = df[['title', 'imdb_score']].drop_duplicates()

# Sort and get top 10 highest rated shows
top10 = df_unique.sort_values(by='imdb_score', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.barh(top10['title'], top10['imdb_score'])
plt.xlabel('IMDb Rating')
plt.ylabel('Title')
plt.title('Top 10 Highest Rated Shows')
plt.gca().invert_yaxis()  # Highest rating on top
plt.show()
```
